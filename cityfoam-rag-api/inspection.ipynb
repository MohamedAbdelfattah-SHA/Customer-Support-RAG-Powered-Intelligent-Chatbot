```python
import chromadb
from src.config import Config

# Connect to your persistent database folder
client = chromadb.PersistentClient(path=Config.CHROMA_DB_DIR)

try:
    collection = client.get_collection(name=Config.COLLECTION_NAME)
    
    # Retrieve all items stored in the collection
    all_data = collection.get(include=["metadatas", "documents"])
    metadatas = all_data["metadatas"]
    documents = all_data["documents"]
    
    # Dictionary to hold our counts
    file_counts = {}
    
    # Loop through metadata to group chunks by their source filename
    for meta in metadatas:
        source_file = meta.get("source", "Unknown File")
        file_counts[source_file] = file_counts.get(source_file, 0) + 1
        
    print("\n" + "="*50)
    print("      DATABASE INGESTION CHUNK ANALYSIS")
    print("="*50)
    print(f"Total Chunks in ChromaDB: {len(metadatas)}\n")
    print("Breakdown per file:")
    print("-" * 50)
    
    for filename, count in file_counts.items():
        print(f" 📂 {filename:<30} -> Split into {count} chunks/parts")
        
    print("="*50 + "\n")

except Exception as e:
    print(f"Error reading database: {e}. Ensure you ran ingest.py first.")
```
